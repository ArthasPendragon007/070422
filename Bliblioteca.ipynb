```python
class Livro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor
        self.disponivel = True
#--------------------------------------------------------------------------------------
    def emprestar(self):
        if self.disponivel:
            self.disponivel = False
            return True
        return False
#--------------------------------------------------------------------------------------
    def devolver(self):
        self.disponivel = True
#--------------------------------------------------------------------------------------
    def __str__(self):
        return f"{self.titulo} por {self.autor} - {'Disponível' if self.disponivel else 'Emprestado'}"


class Usuario:
    def __init__(self, nome):
        self.nome = nome
        self.livros_emprestados = []
#--------------------------------------------------------------------------------------
    def pegar_emprestado(self, livro):
        if livro.emprestar():
            self.livros_emprestados.append(livro)
            return True
        return False
#--------------------------------------------------------------------------------------
    def devolver_livro(self, livro):
        if livro in self.livros_emprestados:
            livro.devolver()
            self.livros_emprestados.remove(livro)
            return True
        return False
#--------------------------------------------------------------------------------------
    def __str__(self):
        livros = ', '.join([livro.titulo for livro in self.livros_emprestados])
        return f"{self.nome} - Livros emprestados: {livros if livros else 'Nenhum'}"
#--------------------------------------------------------------------------------------

class Biblioteca:
    def __init__(self):
        self.livros = []
        self.usuarios = []
#--------------------------------------------------------------------------------------
    def adicionar_livro(self, livro):
        self.livros.append(livro)
#--------------------------------------------------------------------------------------
    def adicionar_usuario(self, usuario):
        self.usuarios.append(usuario)
#--------------------------------------------------------------------------------------
    def remover_livro(self, titulo_livro, autor_livro):
        livro = next((livro for livro in self.livros if livro.titulo == titulo_livro and livro.autor == autor_livro), None)

        if livro is not None:
            # Remove o livro da lista
            self.livros.remove(livro)
            print(f"Livro '{livro.titulo}' removido da biblioteca.")
            print("--"*43)
            print(f"Livros disponíveis após a remoção de '{livro.titulo}':")
            biblioteca.listar_livros()
            print("--"*43)
        else:
            print(f"Livro com título '{titulo_livro}' e autor '{autor_livro}' não encontrado na biblioteca.")
#--------------------------------------------------------------------------------------
    def emprestar_livro(self, titulo_livro, nome_usuario):
        livro = next((livro for livro in self.livros if livro.titulo == titulo_livro), None)
        usuario = next((usuario for usuario in self.usuarios if usuario.nome == nome_usuario), None)

        if livro and usuario:
            if usuario.pegar_emprestado(livro):
                print(f"Livro '{titulo_livro}' emprestado para {nome_usuario}.")
            else:
                print(f"Livro '{titulo_livro}' não está disponível.")
        else:
            print("Livro ou usuário não encontrado.")
#--------------------------------------------------------------------------------------
    def devolver_livro(self, titulo_livro, nome_usuario):
        livro = next((livro for livro in self.livros if livro.titulo == titulo_livro), None)
        usuario = next((usuario for usuario in self.usuarios if usuario.nome == nome_usuario), None)

        if livro and usuario:
            if usuario.devolver_livro(livro):
                print(f"Livro '{titulo_livro}' devolvido por {nome_usuario}.")
            else:
                print(f"Livro '{titulo_livro}' não estava emprestado por {nome_usuario}.")
        else:
            print("Livro ou usuário não encontrado.")
#--------------------------------------------------------------------------------------
    def listar_livros(self):

        if self.livros:
            for livro in self.livros:
                print(livro)
        else:
            print("Nenhum livro disponível na biblioteca.")
#--------------------------------------------------------------------------------------
    def listar_usuarios(self):
        if self.usuarios:
            for usuario in self.usuarios:
                print(usuario)
        else:
            print("Nenhum usuário registrado.")
#--------------------------Fim das classes---------------------------------------------

# Bliblioteca
biblioteca = Biblioteca()

# Adicionar livros
livro1 = Livro("1984", "George Orwell")
livro2 = Livro("Hobbit", "J.R.R. Tolkien")
livro3 = Livro("Fahrenheit 451", "Ray Bradbury")
biblioteca.adicionar_livro(livro1)
biblioteca.adicionar_livro(livro2)
biblioteca.adicionar_livro(livro3)
# Adicionar usuários
usuario1 = Usuario("Alice")
usuario2 = Usuario("Bob")
print("--"*43)

biblioteca.adicionar_usuario(usuario1)
biblioteca.adicionar_usuario(usuario2)

# Lista Pré empréstimos
print("Livros disponíveis antes dos empréstimos:")
biblioteca.listar_livros()
print("--"*43)

# Emprestar livros
biblioteca.emprestar_livro("1984", "Alice")
biblioteca.emprestar_livro("Hobbit", "Bob")
print("--"*43)

# Lista pós emprestimos
print("Livros disponíveis após os empréstimos:")
biblioteca.listar_livros()
print("--"*43)

# Devolver livros
biblioteca.devolver_livro("1984", "Alice")

# Remover o livro "Fahrenheit 451"
biblioteca.remover_livro("Fahrenheit 451", "Ray Bradbury")

# Exibir a lista de livros disponíveis após a remoção
print("Livros disponíveis após a remoção de 'Fahrenheit 451':")
biblioteca.listar_livros()
print("--"*43)

#Inf usuario
print("Informações dos usuários:")
biblioteca.listar_usuarios()
```
